```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
stu_df=pd.read_csv('StudentsPerformance.csv')
```

```python
print(stu_df.head(5)) 
```

```python
stu_df.describe()
```

```python
stu_df.drop(['lunch','parental level of education'],axis=1,inplace=True)
print(stu_df.head(5)) 
```

```python
stu_df['race/ethnicity'].unique()
stu_df['race/ethnicity']=stu_df['race/ethnicity'].map({
    'group A':'Asian Students',
    'group B':'‘African Students',
    'group C':'Afro-Asian Students',
    'group D':'‘American Students',
    'group E':'‘European Students'
})
print(stu_df.head(5)) 
```

```python
ax=sns.countplot(x='test preparation course',hue='gender',data=stu_df)
ax.set(title='Number of Male & Female students who took up the \
‘test preparation course’ and those who did not', ylabel='No of students')
plt.show()
```

```python
ax=sns.countplot(hue='race/ethnicity',x='gender',data=stu_df)
ax.set(title='Number of Male & Female Students belonging \
       to each student group', ylabel='No of students')
plt.show()
```

```python
intervals=(0,40,50,75,100)
categories=['Fail','Second class','First class','Distinction']
```

```python
stu_df['marks_cat']=pd.cut(stu_df.mathscore,intervals,labels=categories)
ax=sns.countplot(x='marks_cat',data=stu_df)
ax.set(title="Maths",xlabel='Category',ylabel='No of students')
plt.show()
```

```python
stu_df['marks_cat']=pd.cut(stu_df['reading score'],intervals,labels=categories)
ax=sns.countplot(x='marks_cat',data=stu_df)
ax.set(title="Readng",xlabel='Category',ylabel='No of students')
plt.show()
```

```python
stu_df['marks_cat']=pd.cut(stu_df['writing score'],intervals,labels=categories)
ax=sns.countplot(x='marks_cat',data=stu_df)
ax.set(title="Writing",xlabel='Category',ylabel='No of students')
plt.show()
```
